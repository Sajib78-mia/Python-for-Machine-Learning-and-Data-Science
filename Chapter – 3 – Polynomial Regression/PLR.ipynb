# Data Preprocessing

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[ : , 1:-1 ].values
y = dataset.iloc[ : , -1 ].values
```

## Training the simple linear regression model on the training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
```

## Predicting the linear regression results

```python
predicted_salary = regressor.predict(X)
```

```python
print(predicted_salary)
```

## Visualizing the linear regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Level VS Salary')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Creating polynomial features of the independent variable

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
```

```python
print(X_poly)
```

## Training the Polynomial Regression Model on the Whole Dataset

```python
regressor_2 = LinearRegression()
regressor_2.fit(X_poly, y)
```

## Predicting the polynomial regression results

```python
predicted_salary_2 = regressor_2.predict(X_poly)
```

```python
print(y)
```

```python
print(predicted_salary_2)
```

## Visualizing the polynomial regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, regressor_2.predict(X_poly), color = 'blue')
plt.title('Level VS Salary')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result with polynomal regression

```python
print(regressor_2.predict(poly_reg.fit_transform([[6.5]])))
```
